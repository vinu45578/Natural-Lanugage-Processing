# **nltk** - special tool kit designed for the NLP

```python
import nltk
nltk.download('punkt_tab')      #downloading a small model i.e punkt_tab. This is for English language (there are other models too)
```

Breakdown of sentence into individual words - **Tokenization**

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentence = "Hello, I'm Nayana. How have you been?"
print(word_tokenize(sentence))       #this is word tokenization
print(sent_tokenize(sentence))       #this is sentence tokenization
# This is in a simplified form,this can be written in more steps by assigning "word_tokenization(sentence)" to some other variable and then print that variable. Same for sentence tokenization.
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

input= "I am tired. I need rest. "
doc = nlp(input)                          #this line represents the nlp pipeline
```

```python
#for word
for token in doc:
  print(token)
```

```python
#for sentence
for token in doc.sents:
  print(token)
```
