```python
#for the pdf -- need a library called PyPDF2
import PyPDF2
pdf_file=open('/content/2494_Nayana_S_P_Offer_Letter (1).pdf','rb')
reader=PyPDF2.PdfReader(pdf_file)
text=""
for page in reader.pages:
  text+=page.extract_text()
print(text)
pdf_file.close()
```
