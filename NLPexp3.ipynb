```python
#some operation with RE- going to remove space in text
pattern = r'\s'
input['twitts'].apply(lambda x:re.sub(pattern,'',x))
```

```python
#here new column is created to find the length of the charecters
pattern = r'\s'
input['Char_count'] = input['twitts'].apply(lambda x:len(re.sub(pattern,'',x)))
input
```

```python
input['Word_count'] = input['twitts'].apply(lambda x:len(x.split()))
input
```

```python
input['avg_word_length'] = input['Char_count']/input['Word_count']
input
```
